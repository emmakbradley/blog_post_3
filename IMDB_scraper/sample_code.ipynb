```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("results.csv")
df = df.dropna()
```

```python
all_names = df['movie_or_tv_name']
print(len(all_names))
all_names = list(set(all_names))
print(len(all_names))
```

```python
# get all unique movie or tv show names
all_names = df['movie_or_tv_name']
unique_list = []
for names in all_names:
    shorter_names = names.split(",")
    unique_list += shorter_names
    
unique_list = list(set(unique_list))

'''
len(unique_list)
unique_np = np.array(unique_list)
unique_np = np.unique(unique_np)
unique_np = unique_np.tolist()
unique_np'''
```

```python
mydict = {}
for name in unique_np:
    this = df['movie_or_tv_name'].str.contains(name)
    count = sum(this)
    mydict[name] = count
mydict
```

```python
mydf = pd.DataFrame(mydict.items(), columns=['movie_or_tv_name', 'count'])
```
