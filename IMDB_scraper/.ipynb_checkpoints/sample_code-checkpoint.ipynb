```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("results.csv")
df = df.dropna()
```

```python
df
```

```python
# get all unique movie or tv show names
all_names = df['movie_or_tv_name']
unique_list = []
for names in all_names:
    shorter_names = names.split(",")
    unique_list += shorter_names

len(unique_list)
unique_np = np.array(unique_list)
unique_np = np.unique(unique_np)
```

```python
df = pd.read_csv("results.csv")
df = pd.concat([df['actor'], df['movie_or_tv_name'].str.split(', ', expand=True)], axis=1)
df
```

```python
df = pd.read_csv("results.csv")
all_titles = df['movie_or_tv_name'].str.split(',', expand=True)
df = df.join(all_titles)
df = df.drop(['movie_or_tv_name'], axis=1)
df
```

```python
df = pd.read_csv("results.csv")
mydict = {}
for i in range(len(unique_np)):
    name = unique_np[i]
    this = df['movie_or_tv_name'].str.contains('Station 19')
    count = sum(this)
    mydict[name] = count
mydict
```
